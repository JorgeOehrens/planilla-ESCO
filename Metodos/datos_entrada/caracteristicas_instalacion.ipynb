```python
## detalles_panel()
#potencia_calculada
#costo_total_anual_om


## inversores ()
#potencia_inversores
#combinacion_inversores

##detalles_panel()
#cantidad_paneles
#potencia_calculada


##generacion_especifica()
#gen_especi
##generacion_anual()
#generacion_anual

##Informacion_energetica()
#total kwH/año



########DATA
# costo_clp_wp_comision=621.8



###### INPUT ######
#potencia_deseada



def caracteristica_instalacion(potencia_deseada: float ,potencia_inversores:float,
                                potencia_calculada : float,
                                combinacion_inversores:str, cantidad_paneles:float ,comuna:str, 
                                generacion_especifica:float , generacion_anual:float, 
                                total_kwh:float, categoria_tipo_montaje:str, 
                                costo_total_anual_om:float, costo_clp_wp_comision: float,
                                costos_adicionales:float ) -> dict:
    potencia_peak_deseada=potencia_deseada
    

    sobredimensionamiento = potencia_calculada / potencia_inversores
    potencia_inversores = potencia_inversores
    combinacion_inversores = combinacion_inversores
    cantidad_paneles = cantidad_paneles
    potencia_peak = potencia_calculada
    comuna_instalacion = comuna

    if comuna_instalacion == "Alto del Carmen":
        region_especifica =  "Region de Atacama"
    else:
        region_especifica = "Region de Atacama"
    
    generacion_especifica = generacion_especifica

    generacion_inicial_total = generacion_anual
    balance_generacion = generacion_inicial_total/total_kwh
    categoria_zona = region_especifica
    categoria_tipo_montaje = categoria_tipo_montaje
    
    if categoria_tipo_montaje == "Suelo":
        superficie_max_paneles = potencia_peak * 13

    elif categoria_tipo_montaje == "Techo estructura inclinada":
        superficie_max_paneles = potencia_peak * 13
    
    else:

        superficie_max_paneles = potencia_peak * 9
    
    costo_total_anual_om = costo_total_anual_om
    costo_planta_solar =  (potencia_peak * 1000 * costo_clp_wp_comision)
    costos_adicionales = costos_adicionales

    costo_total = costo_planta_solar + costos_adicionales


    detalle_caracteristica_instalacion = []
    detalle_caracteristica_instalacion.append ({
        "Potencia peak deseada" : potencia_peak_deseada,
        "Sobredimensionamiento" : sobredimensionamiento,
        "Potencia inversores" : potencia_inversores,
        "Combinación de inversores" : combinacion_inversores,
        "Potencia de panel" : 0,
        "Cantidad de paneles" : cantidad_paneles,
        "Potencia peak" : potencia_peak, 
        "Comuna instalacion" : comuna_instalacion,
        "Region/Zona" :  region_especifica,
        "Generacion especifica" : generacion_especifica,
        "Generacion inicial total" : generacion_inicial_total,
        "Balance generacion/consumo" : balance_generacion,
        "Categoria Zona" : categoria_zona,
        "Categoria Tipo Montaje" : categoria_tipo_montaje,
        "Superficie máx. paneles" : superficie_max_paneles,
        "Costo anual por O&M" : costo_total_anual_om,
        "Costo Planta Solar" : costo_planta_solar,
        "Costo adicionales" : costos_adicionales,
        "Costo total" : costo_total

    })

    return detalle_caracteristica_instalacion
```
