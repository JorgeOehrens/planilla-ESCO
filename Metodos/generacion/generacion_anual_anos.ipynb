```python
##geracion_anual()
# generacion_anual


##detalles_panel()
#potencia_calculada

#DATA
#caracteristicas_panel
def generacion_anual_anos(generacion_anual:float , potencia_calculada:float, caracteristicas_panel:int):
    generacion_anual_anos_data= []

    if caracteristicas_panel == 540:
        degradacion_anual=0.0045

    for a in range(26):

        if a == 0:
            factor_degradacion=0
            generacion_anual_anos_info=0
            kwh=0

        elif a == 1:
            factor_degradacion=1
            generacion_anual_anos_info = generacion_anual
            kwh = generacion_anual_anos_info/potencia_calculada

        else:
            factor_degradacion = factor_degradacion-degradacion_anual
            generacion_anual_anos_info=generacion_anual * factor_degradacion

            kwh = generacion_anual_anos_info/potencia_calculada

        generacion_anual_anos_data.append({
            "Año" : a,
            "Factor degradacion": factor_degradacion,
            "Generacion año": generacion_anual_anos_info,
            "kWh/kWp": kwh,
        })

    return generacion_anual_anos_data
```
